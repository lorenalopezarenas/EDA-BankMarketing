Variables categóricas:

- `job`, `marital`, `education`: múltiples categorías, algunas muy concentradas en ciertos valores.

Variable objetivo:

- `y` está fuertemente desbalanceada, con predominio de *no*.

Variables temporales

- `date`: aparece como variable categórica, por lo que habría que transformarla. También se observa que contiene un gran número de valores únicos, lo que indica alta variabilidad temporal y la necesidad de transformarla en variables más informativas (mes, año).


💡 **Conclusión:**

El Dataset requiere limpieza y estandarización antes del análisis, debido a valores nulos, inconsistencias y desbalances. Abordar estas cuestiones permitirá generar insights precisos y confiables.
